```python
first_side = float(input("enter the first side size of triangle:"))
second_side = float(input("enter the second side size of triangle:"))
third_side = float(input("enter the third side size of triangle:"))

if (first_side+second_side>third_side) and (first_side<second_side+third_side) and (first_side+third_side>second_side):
    print("This is triangle")
else :
    print("This is not a triangle")
```

```python
loop = 1
while loop == 1:
    num1 = float(input("Enter number 1: "))
    operator = input("Enter the operator (+, -, *, /): ")
    num2 = float(input("Enter number 2: "))
    answer = 0.0
    if operator == "+":
        answer = num1 + num2
    elif operator == "-":
        answer = num1 - num2
    elif operator == "*":
        answer = num1 * num2
    elif operator == "/":
        if num2 != 0:
            answer = num1 / num2
        else:
            answer = "Division by zero Error"
    else:
        print("Invalid operator!!!!!")
        continue
    print(f"Result: {answer}")
    choice = input("Do you want to calculate more? (yes/no): ").lower()
    if choice != "yes":
        loop = 0
        print("Goodbye!")
```

```python
w = float(input("Enter your weight in kg: "))
h = float(input("Enter your height in meters: "))

bmi = w/ (h*h)
if bmi < 18:
    result = "Underweight"
elif 18 <= bmi <= 25:
    result = "Normal"
else:
    result = "Overweight"

print(f"Your BMI is: {bmi:.2f}")
print(f"Classification: {result}")
```

```python
name = input("Enter student name: ")

subject1 = float(input("Enter marks for Subject 1: "))
subject2 = float(input("Enter marks for Subject 2: "))
subject3 = float(input("Enter marks for Subject 3: "))

total_marks = subject1 + subject2 + subject3
percentage = (total_marks / 300) * 100

if percentage >= 75:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 45:
    grade = "C"
elif percentage >= 30:
    grade = "D"
else:
    grade = "F"

print("\n" + "="*22)
print("  STUDENT MARK SHEET")
print("="*20)
print(f"Name:       {name}")
print(f"Total:      {total_marks}/300")
print(f"Percentage: {percentage:.2f}%")
print(f"Grade:      {grade}")
print("="*20)
```
